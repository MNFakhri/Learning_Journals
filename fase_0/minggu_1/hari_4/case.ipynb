### creating colums

```python
data_frame["selisih_poin"] = data_frame["pts"] - data_frame["opp_pts"]
data_frame.head()
```

pake looping
- siapkan list kosong sebagai tempat tampung
- siapkan variabel yg dibutuhkan
- buat prosedur
- hasilnya simpan di list kosong
- loop selesai dan list penuh > kolom baru

```python
# list kosong

hasil = []

# kolom pts dan kolom oppts

for i in range(0,len(data_frame)):
    selisih = data_frame.loc[i,"pts"] - data_frame.loc[i,"opp_pts"]
    hasil.append(selisih)

data_frame["selisih_loop"] = hasil
```

```python
# kolom baru hasil pertandingan
# def fucntion

def hasil(selisih):
    if selisih > 0:
        return "menang"
    elif selisih < 0:
        return "kalah"
    else:
        return "seri"
```

```python
hasil_tanding = []

for i in range (0, len(data_frame)):
    result = hasil(data_frame.iloc[i,-2])
    hasil_tanding.append(result)

data_frame["hasil"] = hasil_tanding
```
