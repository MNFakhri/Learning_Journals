Buatlah sebuah kode untuk mensimulasikan melempar koin sebanyak 50.000 kali dengan assign value
random berupa 0 dan 1 untuk kepala atau ekor. Setiap kali koin dilempar, peluang untuk mendapatkan
kepala atau ekor adalah 50%.

```python
import random
from matplotlib import pyplot as plt
```

```python
heads_tails = [0,0]
percobaan_awal = 0
percobaan_akhir = 50000
while percobaan_awal < percobaan_akhir:
    percobaan_awal = percobaan_awal + 1
    pelemparan = random.randint(0,1)
    heads_tails[pelemparan] = heads_tails[pelemparan] + 1
print (heads_tails)
plt.figure(figsize=(5,5))
plt.pie(heads_tails, labels=['heads', 'tails'])
plt.legend()
plt.show()
```

```python
depan_belakang = [0,0]
coba_awal = 0
coba_akhir = 50000
for coba_awal in range(coba_awal,coba_akhir):
    lempar = random.randint(0,1)
    depan_belakang[lempar] = depan_belakang[lempar] + 1
print(depan_belakang)
```

```python
plt.figure(figsize=(10,10))
plt.pie(depan_belakang, labels=["Depan","Belakang"])
plt.legend()
plt.show()
```
