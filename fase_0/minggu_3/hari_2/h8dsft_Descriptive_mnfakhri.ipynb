```python
import numpy as np
import pandas as pd
```

```python
data_covid = pd.read_csv("covid_19_indonesia_time_series_all.csv")
pd.set_option("display.max.columns", None)
data_covid
```

```python
data_covid.info()
```

```python
data_covid.describe()
```

Terdapat data yang terdiri dari number dan object maka harus dipisahkan terlebih dahulu antara variabel berisi angka dan variabel berisi object

```python
data_angka = data_covid.select_dtypes(include="number").columns.tolist()
data_objek = data_covid.select_dtypes(include="object").columns.tolist()
print(data_angka)
print(data_objek)
```

Menghitung rata-rata

```python
rata_rata = data_covid[data_angka].mean()
rata_rata
```

Mencari nilai median

```python
median = data_covid[data_angka].median()
median
```

Mencari nilai Modus

```python
modus_data_angka = data_covid[data_angka].mode()
modus_data_angka
```

```python
modus_data_objek = data_covid[data_objek].mode()
modus_data_objek
```

Menghitung nilai variansi

```python
variansi_data_angka = data_covid[data_angka].var()
variansi_data_angka
```

Menghitung nilai standar deviasi

```python
std_data_angka = data_covid[data_angka].std()
std_data_angka
```
